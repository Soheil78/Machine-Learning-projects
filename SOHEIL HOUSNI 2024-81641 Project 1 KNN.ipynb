```python
#Generation of the dataset and construction of the KNN algorithm

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
import math

np.random.seed(0)

x,y=make_classification(n_samples=100, n_features=2, n_informative=2, n_redundant=0, n_clusters_per_class=1, n_classes=2)

def knn(point,k):
    distance_list=[]
    for i in x:
        distance=math.sqrt((i[0]-point[0])**2+(i[1]-point[1])**2)
        distance_list.append(distance)
    sort=np.argsort(distance_list)
    
    number_0=[]
    number_1=[]
    
    for i in range(k):
        if y[sort[i]]==0:
            number_0.append(y[sort[i]])
        if y[sort[i]]==1:
            number_1.append(y[sort[i]])
    
    if len(number_0)>=len(number_1):
        prediction=0
    elif len(number_0)<len(number_1):
        prediction=1
    return prediction

```

```python
#Results of constructed model with k=5

plt.scatter(x[:,0][y.ravel() == 0], x[:,1][y.ravel() ==0], color='blue', label='Class 0')
plt.scatter(x[:,0][y. ravel() == 1], x[:,1][y.ravel() ==1],color='red', label='Class 1')
plt.legend()

h = 0.05  
e = 1 
x1_min, x1_max = x[:, 0].min() - e, x[:, 0].max() + e
x2_min, x2_max = x[:, 1].min() - e, x[:, 1].max() + e
xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, h), np.arange(x2_min, x2_max, h))
Z=[]
k=5
for i in (np.c_[xx1.ravel(), xx2.ravel()]):
    Z.append(knn(i,k))
    T=np.array(Z)
T = T.reshape(xx1.shape)

# Plot the decision boundary and data points
plt.contourf(xx1, xx2, T, alpha=0.3, cmap='coolwarm')
plt.scatter(x[:, 0], x[:, 1], c=y, cmap='coolwarm', edgecolors='k')
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('KNN Classification')
plt.show()

```

```python
#Results of Sklearn model with k=5

from sklearn.neighbors import KNeighborsClassifier
knn_sk = KNeighborsClassifier(n_neighbors=5)
knn_sk.fit(x, y)
y_pred = knn_sk.predict(x)

h = 0.05  
e = 1 
x1_min, x1_max = x[:, 0].min() - e, x[:, 0].max() + e
x2_min, x2_max = x[:, 1].min() - e, x[:, 1].max() + e
xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, h), np.arange(x2_min, x2_max, h))
Z = knn_sk.predict(np.c_[xx1.ravel(), xx2.ravel()])
Z = Z.reshape(xx1.shape)

plt.contourf(xx1, xx2, Z, alpha=0.3, cmap='coolwarm')
plt.scatter(x[:, 0], x[:, 1], c=y, cmap='coolwarm', edgecolors='k')
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('KNN Classification (sklearn)')
plt.show()
```
